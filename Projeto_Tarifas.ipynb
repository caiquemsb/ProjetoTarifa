```python
#Ajuste de nomes de colunas

df_202301.rename(columns= {"ANO":"Ano de Referência", "MES":"Mês de Referência", "EMPRESA":"ICAO Empresa Aérea", "ORIGEM":"ICAO Aeródromo Origem", "DESTINO":"ICAO Aeródromo Destino", "TARIFA":"Tarifa-N", "ASSENTOS":"Assentos Comercializados"}, inplace=True)
df_202302.rename(columns= {"ANO":"Ano de Referência", "MES":"Mês de Referência", "EMPRESA":"ICAO Empresa Aérea", "ORIGEM":"ICAO Aeródromo Origem", "DESTINO":"ICAO Aeródromo Destino", "TARIFA":"Tarifa-N", "ASSENTOS":"Assentos Comercializados"}, inplace=True)
df_202303.rename(columns= {"ANO":"Ano de Referência", "MES":"Mês de Referência", "EMPRESA":"ICAO Empresa Aérea", "ORIGEM":"ICAO Aeródromo Origem", "DESTINO":"ICAO Aeródromo Destino", "TARIFA":"Tarifa-N", "ASSENTOS":"Assentos Comercializados"}, inplace=True)
```

```python
distances = df_trimestre1_geral_3.apply(
    lambda row: distance.distance(
        (row["Latitude - Origem"], row["Longitude - Origem"]), (row["Latitude - Destino"], row["Longitude - Destino"])
    ), axis=1
)
```

```python
distances
```

```python
df_distances = pd.DataFrame(distances)
```

```python
df_distances
```

```python
df_trimestre1_geral_3["Distância em km"] = df_distances
```

```python
df_trimestre1_geral_3
```

```python
df_trimestre1_geral_3["Trecho"] = df_trimestre1_geral_3["Cidade - Origem"] + "/" + df_trimestre1_geral_3["Estado - Origem"] + " - " + df_trimestre1_geral_3["Cidade - Destino"] + "/" + df_trimestre1_geral_3["Estado - Destino"]
```

```python
#https://stackoverflow.com/questions/49274833/how-do-i-convert-vincenty-distance-to-float-in-pandas-dataframe

df_trimestre1_geral_3['Distância entre trechos em km'] = df_trimestre1_geral_3['Distância em km'].astype(str).str[:-3].astype(float)
```

```python
df_trimestre1_geral_3["Tarifa por km"] = df_trimestre1_geral_3["Tarifa-N"]/df_trimestre1_geral_3["Distância entre trechos em km"]
```

```python
df_trimestre1_geral_3
```
